```python
from Landsat8.landsat8_utils import create_raster_stack
from osgeo import gdal
from Landsat8.spectral_vegatation_indices.VCI import calculate_VCI, plot_VCI, classify_VCI
from utils import splitall
import os
import numpy as np
```

```python
vci_data = {}
```

```python
data_path = "./data"
kmls_path = data_path + "/kmls"
banadir_path = "./data/banadir"
```

```python
create_raster_stack(banadir_path, "banadir")
os.mkdir(banadir_path + "/raster_stack/clipped/")
```

```python
banadir_raster_path = banadir_path +"/raster_stack"
for (dirpath, dirnames, filenames) in os.walk(banadir_raster_path) :
    for stack in filenames:
        OutTile = gdal.Warp(banadir_path  + "/raster_stack/clipped/" + stack, 
                    banadir_raster_path + "/"+stack, 
                    cutlineDSName=banadir_path + '/banadir.kml',
                    cropToCutline=True,
                    dstNodata = 0)

        OutTile = None
```

```python
banadir_clipped_raster_path = banadir_path +"/raster_stack/clipped"
for (dirpath, dirnames, filenames) in os.walk(banadir_clipped_raster_path) :
    region_vci_data = {}
    for stack in filenames:
        vci = calculate_VCI(banadir_clipped_raster_path + "/" +stack, banadir_clipped_raster_path)
        print(np.average(vci))
        plot_VCI(vci,banadir_clipped_raster_path + "/" +stack)
        #classify_VCI(vci,banadir_clipped_raster_path + "/" +stack)
```

```python


vci_data = {}
for (dirpath, dirnames, filenames) in os.walk("./data/") :
    for region in dirnames :
        region_path = "./data/" + region
        kml_path = region_path + "/" + region + ".kml"
        create_raster_stack(region_path, region)
        os.mkdir(region_path + "/raster_stack/clipped/")

        for (dirpath, dirnames, filenames) in os.walk(region_path + "/raster_stack") :
            for stack in filenames:
                OutTile = gdal.Warp(region_path  + "/raster_stack/clipped/" + stack, 
                    banadir_raster_path + "/" +stack, 
                    cutlineDSName=kml_path,
                    cropToCutline=True,
                    dstNodata = 0)

                OutTile = None
        
        for (dirpath, dirnames, filenames) in os.walk(region_path  + "/raster_stack/clipped/") :
            region_vci_data = {}
            for stack in filenames:
                vci = calculate_VCI(region_path  + "/raster_stack/clipped/" +stack, region_path  + "/raster_stack/clipped/")
                region_vci_data[stack.replace(".tif", "").replace(region + "-", "")] = np.average(vci);
        
        vci_data[region] = vci_data
            
```
